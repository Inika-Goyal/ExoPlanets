```python
import pandas as pd

# Load your full Kepler dataset
df = pd.read_csv("kepler_data.csv", comment="#")

# Select the first 10 rows and required columns
planet_csv = df[['kepler_name', 'ra', 'dec', 'kepoi_name', 'koi_sma', 'koi_steff']].head(10)

# Rename columns for clarity
planet_csv = planet_csv.rename(columns={
    'kepler_name': 'planet_name',
    'kepoi_name': 'host_star',
    'koi_sma': 'distance_from_star_au',
    'koi_steff': 'stellar_temp_K'
})

# Save to CSV
planet_csv.to_csv("planets_sample.csv", index=False)

print(planet_csv)
```
